```python
#bar chart
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df
```

```python
df=df.groupby('who')['fare'].sum().to_frame().reset_index()
df
```

```python
#bar chart using matplotlib
plt.barh(df['who'],df['fare'],color=['blue','black','red'])
plt.title('bar chart')
plt.xlabel('fare')
plt.ylabel('who')
plt.show()
```

```python
#bar chart using seaborn
sns.barplot(x = 'fare',y = 'who',data = df ,palette = "Blues")
plt.title('Chart title')
plt.xlabel('X axis title')
plt.ylabel('Y axis title')
# Show the plot
plt.show()
```

```python
#column chart using matplotlib
df=sns.load_dataset('Titanic')
df=df.groupby('who')['fare'].sum().to_frame().reset_index()
df
```

```python
plt.bar(df['who'],df['fare'],color=['blue','red','green'])
plt.title('column chart')
plt.xlabel('who')
plt.ylabel('fare')
plt.show()
```

```python
sns.barplot(x='who',y='fare',data=df , palette="Blues")
plt.title('column chart')
plt.xlabel('who')
plt.ylabel('fare')
plt.show()
```

```python
import pandas as pd
import numpy as np
```

```python
#grouped bar chart using matplotlib
df=sns.load_dataset("titanic")
df_pivot=pd.pivot_table(df,values="fare",index="who",columns="class",aggfunc=np.mean)
```

```python
ax=df_pivot.plot(kind="bar", alpha=0.5)
plt.title('grouped bar chart')
plt.xlabel('who')
plt.ylabel('fare')
plt.show()
```

```python
#grouped bar chart using seaborn
sns.barplot(x='who',y='fare' ,data=df ,palette="Blues", hue='class')
plt.title('column chart')
plt.xlabel('who')
plt.ylabel('fare')
plt.show()
```

```python
#stacked bar chart using matplotlib
#creating dataset
df=pd.DataFrame(columns=["A","B","C","D"],data=[["E",0,1,1],
                                              ["F",1,1,0],
                                              ["G",0,1,0]])
df
```

```python
df.plot.bar(x='A',y=["B","C","D"],stacked=True, width=0.4,alpha=0.5)
plt.title("Stacked bar chart")
plt.xlabel("A")
plt.ylabel("values")
plt.show()
```

```python
#usning seaborn 
df=pd.DataFrame(columns=["A","B","C","D"],data=[["E",1,0,1],
                                               ["F",0,1,1],
                                                ["G",1,1,0]])
df
```

```python
df.set_index('A').T.plot(kind='bar',stacked=True)
plt.title("Stacked bar chart")
plt.xlabel("A")
plt.ylabel("values")
plt.show()
```

```python
#line chart using matplotlib
df=sns.load_dataset("iris")
df=df.groupby('sepal_length')['sepal_width'].sum().to_frame().reset_index()
plt.plot(df['sepal_length'],df['sepal_width'])
plt.title("line chart")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#pie chart using matplotlib
cars=['Audi','BMW','Hyndai','Honda']
data=[20,15,15,14,]
plt.pie(data, labels=cars,colors= ['blue','red','pink','black'])
plt.title(' pie chart')
plt.show()
```

```python
#area chart using matplotlib
import matplotlib.pyplot as plt
x=range(1,6)
y=[[1,4,6,8,9], [2,2,7,10,12], [2,8,5,10,6]]
#Creating the area chart
ax = plt.gca()
ax.stackplot(x, y, labels=['A','B','C'],alpha=0.5)
#Adding the aesthetics
plt.legend(loc='upper left')
plt.title('Chart title')
plt.xlabel('X axis title')
plt.ylabel('Y axis title')
#Show the plot
plt.show()
```

```python
#area chart using seaborn
year=[1,2,3]
salary=[[2,4,6],[4,6,8],[3,6,9]]
plt.stackplot(year,salary, labels=['A','B','C'])
plt.title('Chart title')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
